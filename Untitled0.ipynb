## Analyze Review Trends Over Time

### Subtask:
Examine how the number of reviews and average ratings change over time using the 'Review Date' column, plotting the trends with a legend.


**Reasoning**:
To prepare the 'Review Date' column for time-series analysis, I will convert it to datetime objects and then extract only the date part to create a new 'Review_Day' column, as specified in the instructions.



```python
df['Review Date'] = pd.to_datetime(df['Review Date'])
df['Review_Day'] = df['Review Date'].dt.date
print("Converted 'Review Date' to datetime and created 'Review_Day' column.")
```

**Reasoning**:
To analyze trends over time, I will group the DataFrame by the 'Review_Day' column, calculate the count of reviews and the average rating for each day, and store these in a new DataFrame for plotting.



```python
daily_trends = df.groupby('Review_Day').agg(
    number_of_reviews=('Review Id', 'size'),
    average_rating=('Ratings', 'mean')
).reset_index()

print("Created daily_trends DataFrame with number of reviews and average ratings per day.")
print(daily_trends.head())
```

**Reasoning**:
To visualize the trends, I will create two subplots: one for the number of reviews over time and another for the average rating over time, ensuring both plots have appropriate titles, labels, and legends.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
plt.figure(figsize=(15, 6))

# Subplot 1: Number of Reviews over time
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.lineplot(x='Review_Day', y='number_of_reviews', data=daily_trends, label='Number of Reviews')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.legend()
plt.grid(True)

# Subplot 2: Average Rating over time
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.lineplot(x='Review_Day', y='average_rating', data=daily_trends, color='orange', label='Average Rating')
plt.title('Average Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("Visualized trends of number of reviews and average ratings over time.")
```
